# Sons musicais: altura, intensidade e timbre

## Altura
Sensação que permite distinguir entre sons mais graves e sons mais agudos.  
A característica associada à altura é a **frequência**: quanto maior a frequência, mais agudo é o som.

## Timbre
É a característica que torna possível diferenciar o som de um violino do som de um violão.  
Está relacionado com as diferentes proporções em que entram os **harmônicos**  
e é determinado pelas propriedades dos materiais.

Uma nota musical pode sempre ser representada como uma **soma de funções seno e cosseno**.

# Ondas sonoras em um gás  
**Pressão (P) vs Deslocamento (u)**

## Relação pressão–deslocamento
$$
\Delta F = A\,[P(x,t) - P(x+\Delta x,t)]
        = -A\,\Delta x\,\frac{\partial P}{\partial x}
$$

$$
\Rightarrow \frac{\partial P}{\partial x}
= -\rho\,\frac{\partial^2 u}{\partial t^2}
$$

## Relação deslocamento–densidade
$$
\Delta V = A\,[u(x+\Delta x,t) - u(x,t)]
        = A\,\Delta x\,\frac{\partial u}{\partial x}
$$

$$
\Rightarrow \frac{\partial u}{\partial x}
= \frac{\Delta V}{V}
= -\frac{\Delta \rho}{\rho}
$$

## Relação densidade–pressão
$$
p = P(x,t) - P_0
\approx \Delta p \left(\frac{\partial P}{\partial \rho}\right)_0
$$

$$
p = -\rho \left(\frac{\partial P}{\partial \rho}\right)_0
\frac{\partial u}{\partial x}
$$

Definindo:
$$
Y = -\frac{\Delta P}{\Delta V / V}
$$

temos:
$$
p = -Y\,\frac{\partial u}{\partial x}
$$

---

## Equação da onda sonora
$$
\frac{\partial^2 p}{\partial x^2}
= Y\,\frac{\partial^2 u}{\partial x^2}
= \rho\,\frac{\partial^2 u}{\partial t^2}
$$

ou:
$$
\frac{\partial^2 u}{\partial x^2}
- \frac{\rho}{Y}\,\frac{\partial^2 u}{\partial t^2} = 0
$$

com velocidade:
$$
v = \sqrt{\frac{Y}{\rho}}
$$

---

## Nós e antinós
- Um **nó de deslocamento** é um **antinó de pressão**: ocorrem as maiores variações de densidade.  
- Um **antinó de deslocamento** é um **nó de pressão**: ocorrem as menores variações de densidade.

# Ondas em instrumentos de sopro

## Tubo aberto

### (a) Fundamental
$$
f_1 = \frac{v}{2L}
$$

- N: nó de deslocamento  
- A: antinó de deslocamento

### (b) Segundo harmônico
$$
f_2 = \frac{2v}{2L}
$$

### (c) n-ésimo harmônico
$$
f_n = \frac{n v}{2L}
\quad\text{com}\quad
L = \frac{n\lambda}{2}
$$

---

## Tubo fechado

### (a) Fundamental
$$
f_1 = \frac{v}{4L}
\quad\text{com}\quad
L = \frac{\lambda}{4}
$$

### (b) Terceiro harmônico
$$
f_3 = \frac{3v}{4L} = 3f_1
\quad\text{com}\quad
L = \frac{3\lambda}{4}
$$

### (c) n-ésimo harmônico
$$
f_n = \frac{n v}{4L}
\quad\text{com}\quad
n \in \mathbb{N},\; n \text{ ímpar}
$$

$$
\lambda_n = \frac{4L}{2n-1}
$$

# Notas em cordas

## Fundamental
$$
\frac{\lambda}{2} = L
$$

$$
\lambda_n = \frac{2L}{n}
$$

$$
f_n = n f_1
\quad\text{com}\quad
f_n = \frac{v}{\lambda_n}
$$
