# Mecânica quântica, estatística e o sólido de Einstein

## Processos Reversíveis

As leis de Newton valem tanto para  
“o tempo indo para frente” quanto para  
“o tempo indo para trás”.

As chamadas forças fundamentais (forte, fraca, eletromagnética e gravitacional)  
respeitam essa **simetria temporal**.

Essa simetria temporal tem relação direta com a **conservação da energia total**.

---

## Fórmula geral para a distribuição de quanta entre os osciladores

$$
\Omega = \frac{(q + N - 1)!}{q!\,(N - 1)!}
$$

Número de maneiras de distribuir $q$ quanta entre $N$ osciladores.

---

## Macroestado e Microestado

**Macroestado:** quanto de quanta está em cada corpo.  
**Microestado:** forma em que o quanta está distribuído.

### Suposição fundamental da Mecânica Estatística

Um sistema de tamanho dado (composto de $N$ osciladores) e energia total fixa  
(número $q$ de quanta) define um **macroestado**.

Uma maneira compatível de distribuir a energia define um **microestado**.

**Suposição fundamental:**  
todos os microestados compatíveis com um dado macroestado são **igualmente prováveis**.

---

## Energia de um Quantum

Frequência natural do oscilador:

$$
\omega_0 = \sqrt{\frac{K_{mi}}{m_{\text{atômica}}}}
$$

Energia do nível $n$:

$$
E_n = \left(n + \frac{1}{2}\right)\hbar \omega_0
$$

onde $\hbar$ é a constante de Planck reduzida e $n$ é o número quântico.

---

## Entropia ($S$)

A entropia do sistema é:

$$
S = k_B \ln(\Omega)
$$

onde

$$
k_B = 1{,}38 \times 10^{-23}\ \text{J/K}
$$

(constante de Boltzmann).

---

## Condição de máximo de entropia

O máximo ocorre quando:

$$
\frac{dS_1}{dq_1} = \frac{dS_2}{dq_2}
$$

A configuração mais provável é aquela que **maximiza a entropia do sistema**  
e é a que melhor distribui a energia entre seus subsistemas.

Essa configuração caracteriza o **equilíbrio térmico**.

---

## Segunda Lei da Termodinâmica

Se um sistema fechado não está em equilíbrio,  
a consequência mais provável é que a **entropia aumente**.

> Desvios são observáveis em pequenas escalas,  
> porém virtualmente impossíveis em objetos macroscópicos.

---

## Temperatura

$$
\frac{dS}{dq_i} \propto \frac{1}{T_i}
$$

Definição:

$$
\frac{1}{T} = \frac{dS}{dE_{\text{int}}}
$$

Como $E_{\text{int}} = q\,\hbar \omega_0$, temos:

$$
\frac{1}{T} = \frac{\Delta S}{\Delta E}
$$

ou

$$
T = \frac{\Delta E}{\Delta S}
$$

---

## Entropia total

Em fenômenos macroscópicos, a entropia total jamais diminui:

$$
\Delta S_{\text{sist}} + \Delta S_{\text{ent}} \ge 0
$$

- Processo **irreversível**:
$$
\Delta S_{\text{sist}} + \Delta S_{\text{ent}} > 0
$$

- Processo **reversível**:
$$
\Delta S_{\text{sist}} + \Delta S_{\text{ent}} = 0
$$

---

## Entropia e Calor

Se a temperatura for constante:

$$
dS = \frac{dQ}{T}
$$

Se a temperatura variar significativamente no processo:

$$
\Delta S = \int_{T_i}^{T_f} \frac{dQ}{T}
$$

Para um sistema com capacidade térmica $C$:

$$
\Delta S = \int_{T_i}^{T_f} \frac{C\,dT}{T}
$$
