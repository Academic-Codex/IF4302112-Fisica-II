## Gases — Escoamento através de um furo (livre caminho livre)

### Densidade de partículas
$$
n = \frac{N}{V}
$$
onde $N$ é o número de partículas e $V$ o volume.

### Caso unidimensional / unidirecional
$$
\frac{N}{\Delta t} = n\,A\,\bar{v}
$$

### Caso tridimensional
$$
\frac{N}{\Delta t} = \frac{1}{4}\,n\,A\,\bar{v}
$$

### Livre caminho médio
$$
\ell \simeq \frac{1}{nA}
$$

com
$$
A \simeq \pi (R + r)^2
$$

## Distribuição de Maxwell–Boltzmann
$$
P(v) = 4\pi \left(\frac{M}{2\pi k_B T}\right)^{3/2} v^2
       e^{-\frac{M v^2}{2 k_B T}}
$$

## Velocidades características

### Velocidade mais provável
$$
v_{mp} = \sqrt{\frac{2 k_B T}{M}}
$$

### Velocidade quadrática média
$$
v_{rms} = \sqrt{\frac{3 k_B T}{M}}
$$

### Velocidade média
$$
\bar{v} = \sqrt{\frac{8 k_B T}{\pi M}}
\;\simeq\; 0{,}92\,v_{rms}
$$

## Equipartição da energia (clássica)
Cada termo quadrático contribui com
$$
\frac{1}{2} k_B T
$$

## Calor específico em gases
$$
E_{\text{molécula}} = c_V\,T
$$

onde
$$
c_V = \frac{n}{2} k_B
$$
e $n$ é o número de graus de liberdade ativos.

## Pressão microscópica de um gás
$$
P = \frac{1}{3}\,n\,m\,\langle v^2 \rangle
$$

Como
$$
\frac{1}{2} m \langle v^2 \rangle = \frac{3}{2} k_B T
$$

segue que
$$
P = n k_B T
$$
