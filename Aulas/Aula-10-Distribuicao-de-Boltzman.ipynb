# Aula 10 — Distribuição de Boltzmann

## Distribuição de Boltzmann

Qual é a probabilidade $P(E)$ de encontrar o sistema de interesse com uma energia $E$?

Primeiramente, supomos que a energia total do sistema é fixa, de modo que

$$
E_{\text{tot}} = E_{\text{res}} + E
$$

A probabilidade $P(E)$ pode ser escrita como

$$
P(E) = \frac{\Omega_{\text{res}}(E_{\text{res}})\,\Omega(E)}{\Omega_{\text{tot}}(E_{\text{tot}})}
$$

ou, equivalentemente,

$$
P(E) = \frac{\Omega_{\text{res}}(E_{\text{tot}} - E)\,\Omega(E)}{\Omega_{\text{tot}}(E_{\text{tot}})}
$$

Aplicando o logaritmo natural e multiplicando por $k_B$, obtemos

$$
k_B \ln P(E)
=
S_{\text{res}}(E_{\text{tot}} - E)
+
k_B \ln \Omega(E)
-
k_B \ln \Omega_{\text{tot}}(E_{\text{tot}})
$$

## Expansão em série de Taylor

Expandindo $S_{\text{res}}(E_{\text{tot}} - E)$ em série de Taylor:

$$
S_{\text{res}}(E_{\text{tot}} - E)
\approx
S_{\text{res}}(E_{\text{tot}})
-
\frac{dS_{\text{res}}}{dE_{\text{res}}}\,E
$$

Como

$$
\frac{dS}{dE} = \frac{1}{T},
$$

segue que

$$
S_{\text{res}}(E_{\text{tot}} - E)
=
S_{\text{res}}(E_{\text{tot}}) - \frac{E}{T}
$$

Substituindo na expressão anterior:

$$
k_B \ln P(E)
=
k_B \ln \Omega(E)
-
\frac{E}{T}
+
\text{constante}
$$

## Forma exponencial

Voltando à forma exponencial:

$$
P(E)
=
A\,\Omega(E)\,e^{-E/(k_B T)}
$$

onde $A$ é uma constante de normalização.

Essa é a **Distribuição de Boltzmann**.

## Distribuição de Maxwell–Boltzmann

A energia total pode ser escrita como

$$
E = E_{\text{trans}} + E_{\text{vib}} + E_{\text{rot}} + U_p + \cdots
$$

Assim, a distribuição de Boltzmann pode ser aplicada a **graus de liberdade específicos**.

### Exemplo — Distribuição de altura em um campo gravitacional

$$
P(y)\,dy \propto e^{-mgy/(k_B T)}\,dy
$$
