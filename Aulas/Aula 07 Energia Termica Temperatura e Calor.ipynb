# Energia térmica, temperatura e calor

## Lei zero da Termodinâmica
Três sistemas $A$, $B$ e $C$ tais que:
- $A$ e $B$ estão em equilíbrio térmico entre si;
- $B$ e $C$ estão em equilíbrio térmico;

então $A$ e $C$ estão em equilíbrio térmico.

---

## Capacidade térmica $(C)$
Energia necessária para elevar a temperatura de um objeto em um grau  
(depende do tamanho do objeto).

Quanto maior a temperatura, maior a capacidade térmica.

$$
C = \frac{\Delta E_{\text{térmica}}}{\Delta T}
$$

---

## Calor específico $(c)$
Capacidade térmica por unidade de massa (mol, átomo, etc.).

$$
c = \frac{C}{m}
= \frac{\Delta E_{\text{térmica}}}{m\,\Delta T}
$$

ou, equivalentemente,

$$
\Delta E_{\text{térmica}} = m\,c\,\Delta T
$$

---

## Calor
Transferência de energia associada a uma diferença de temperatura.

---

# Princípio da energia

Quando há trabalho realizado pelo entorno e transferência de energia
por diferença de temperatura (calor) do entorno para o sistema:

$$
\Delta E_{\text{sistema}} = W_{\text{externo}} + Q
$$

Se:
- $Q > 0$, a energia do sistema **aumenta**;
- $Q < 0$, a energia do sistema **diminui**.

---

## Forma geral do princípio da energia

$$
\Delta E_{\text{sistema}}
= W_{\text{externo}} + Q
+ \text{(outras formas de transferência de energia)}
$$

como, por exemplo:
- transferência de massa,
- ondas mecânicas,
- eletricidade,
- radiação, etc.

---

**Observação:**  
$$
W_{\text{ext}} = -W_{\text{sist}}
$$

onde $W_{\text{ext}}$ é o trabalho que o **entorno atua sobre o sistema**.
