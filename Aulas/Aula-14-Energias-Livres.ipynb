# Entalpia e Energias Livres

## Definição de Entalpia

A **entalpia** é definida a partir da energia interna e das variáveis de estado pressão e volume:

$$
H \equiv E_{\text{int}} + P V
$$

A entalpia é uma **função de estado**, muito útil em processos **isobáricos**.

### Diferencial da entalpia

$$
\Delta H = \Delta E_{\text{int}} + \Delta(PV)
$$

$$
dH = dE_{\text{int}} + d(PV)
$$

Usando a Primeira Lei da Termodinâmica:

$$
dE_{\text{int}} = \delta Q - P\,dV
$$

Logo,

$$
dH = \delta Q - P\,dV + P\,dV + V\,dP
$$

$$
\boxed{dH = \delta Q + V\,dP}
$$

- Se $P$ é constante ($dP = 0$):

$$
\boxed{dH = \delta Q}
$$

---

## Bônus — Relações Termodinâmicas

Sabemos que:

$$
dS = \frac{\delta Q}{T}
$$

Logo,

$$
dH = T\,dS + V\,dP
$$

Considerando $H = H(S,P)$:

$$
dH = \left(\frac{\partial H}{\partial S}\right)_P dS
     + \left(\frac{\partial H}{\partial P}\right)_S dP
$$

Comparando termos:

$$
\boxed{
T = \left(\frac{\partial H}{\partial S}\right)_P
\qquad
V = \left(\frac{\partial H}{\partial P}\right)_S
}
$$

Da igualdade das derivadas mistas:

$$
\frac{\partial^2 H}{\partial P \partial S}
=
\frac{\partial^2 H}{\partial S \partial P}
$$

Obtemos a **relação de Maxwell**:

$$
\boxed{
\left(\frac{\partial T}{\partial P}\right)_S
=
\left(\frac{\partial V}{\partial S}\right)_P
}
$$

---

## Entalpia de um Gás Ideal

Para um gás ideal:

$$
H = E_{\text{int}} + P V = E_{\text{int}} + nRT
$$

Logo,

$$
dH = dE_{\text{int}} + d(nRT)
$$

Se $n$ é constante:

$$
dH = dE_{\text{int}} + nR\,dT
$$

- Se $P$ e $T$ são constantes:

$$
\boxed{dH = 0 \Rightarrow \Delta H = 0}
$$

---

## Variação de Entalpia em Processos Químicos

Para a reação:

$$
A + B \rightarrow C + D
$$

A variação de entalpia é:

$$
\boxed{
\Delta H = (H_C + H_D) - (H_A + H_B)
}
$$

---

## Energias Livres (Funções de Estado)

### Energia Livre de Helmholtz $(F)$

$$
F = E_{\text{int}} - T S
$$

Diferencial:

$$
dF = dE_{\text{int}} - d(TS)
$$

$$
dF = \delta Q + \delta W - S\,dT - T\,dS
$$

Como $\delta Q = T\,dS$:

$$
\boxed{dF = \delta W - S\,dT}
$$

- Em **temperatura constante** (banho térmico):

$$
\boxed{\Delta F = W}
$$

Critério de espontaneidade:

$$
\boxed{\Delta F \le 0}
$$

---

### Energia Livre de Gibbs $(G)$

$$
G = E_{\text{int}} - TS + PV = H - TS
$$

Diferencial:

$$
dG = dH - S\,dT - T\,dS
$$

Usando $dH = T\,dS + V\,dP$:

$$
\boxed{dG = V\,dP - S\,dT}
$$

- Em processos a **pressão e temperatura constantes**:

$$
\boxed{\Delta G = \Delta H - T\Delta S}
$$

Critério de espontaneidade:

$$
\boxed{\Delta G < 0}
$$

---

## Revisão — Entropia em Ciclos

Para um ciclo térmico:

### Entropia do gás

$$
\Delta S_{\text{gás}} = \frac{Q_A}{T_A} - \frac{Q_B}{T_B}
$$

- $Q_A$: calor absorvido (positivo)
- $Q_B$: calor cedido (negativo)

### Entropia do entorno

$$
\Delta S_{\text{entorno}} = -\frac{Q_A}{T_A} + \frac{Q_B}{T_B}
$$

### Entropia do banho térmico

$$
\Delta S_{\text{banho}} = \frac{Q}{T}
$$

Como:

$$
\Delta E_{\text{banho}} = -\Delta E_{\text{sistema}}
$$

então:

$$
\boxed{
\Delta S_{\text{banho}} = -\frac{\Delta E_{\text{sistema}}}{T}
}
$$
