## Princípio da Superposição

Se $f$ e $g$ são funções que satisfazem a equação de onda, então a função
$$
Af + Bg
$$
também satisfaz a equação de onda.

**Obs.:** Quando duas ondas construtivas se encontram na posição de equilíbrio ($u = 0$) e $U$ é máximo, elas se reforçam.  
Se forem de amplitudes opostas, ocorre cancelamento: $u = 0$ e $U = 0$.

## Ondas Estacionárias

No caso de ondas estacionárias, as funções $f$ e $g$ são funções harmônicas, com mesma frequência e mesma amplitude:

$$
u(x,t) = A f(x - vt) + B g(x + vt)
$$

Supondo:
$$
f(x - vt) = \cos(kx - \omega t + \phi_1),
\qquad
g(x + vt) = \cos(kx + \omega t + \phi_2)
$$

Temos:
$$
u = f + g = 2A \cos(kx + \alpha)\cos(\omega t + \beta)
$$

com:
$$
\alpha = \frac{\phi_1 + \phi_2}{2},
\qquad
\beta = \frac{\phi_2 - \phi_1}{2}
$$

## Modos Normais de Vibração

Considerando uma corda vibrante de comprimento $L$, presa em ambas as extremidades, o movimento pode ser descrito por ondas estacionárias.

Condições de contorno:
$$
u(0,t) = u(L,t) = 0
$$

Resulta que:
$$
k_m = \frac{m\pi}{L}, \quad m = 1,2,3,\ldots
$$

Logo:
$$
\lambda_m = \frac{2L}{m}
$$

Ou seja, o comprimento de onda é sempre um submúltiplo de $2L$.

## Reflexão de Ondas

### Extremidade Fixa

Quando a onda atinge uma extremidade fixa, ocorre inversão de fase.

Condição de contorno:
$$
u(0,t) = 0
$$

O que implica:
$$
f(L - vt) + g(L + vt) = 0 \quad \text{(para qualquer $t$)}
$$

---

### Extremidade Livre

Quando a onda é refletida em uma extremidade livre, não há inversão de fase.

Condição de contorno formal:
$$
\frac{\partial u}{\partial x}(0,t) = 0
$$

Logo:
$$
f'(x - vt) + g'(x + vt) = 0
$$

Integrando:
$$
g(\phi) = f(-\phi)
$$

## Batimentos

Interferência de duas ondas harmônicas com frequências muito próximas:

$$
u(x,t) = A\cos(k_1 x - \omega_1 t) + A\cos(k_2 x - \omega_2 t)
$$

Definindo:
$$
k = \frac{k_1 + k_2}{2}, \quad \Delta k = k_1 - k_2
$$
$$
\omega = \frac{\omega_1 + \omega_2}{2}, \quad \Delta \omega = \omega_1 - \omega_2
$$

Obtemos:
$$
u(x,t) =
2A \cos\!\left(\frac{\Delta k\,x - \Delta\omega\,t}{2}\right)
\cos(kx - \omega t)
$$

Velocidade de fase:
$$
v_f = \frac{\omega}{k}
$$

Velocidade de grupo:
$$
v_g = \frac{\Delta\omega}{\Delta k}
\;\;\longrightarrow\;\;
\frac{d\omega}{dk}
$$

## Cone de Mach

Ocorre quando:
$$
v_{\text{obj}} > v_{\text{som}}
$$

Ângulo de Mach:
$$
\sin\alpha = \frac{v_{\text{som}}}{v_{\text{obj}}}
$$

## Interferência (Duas Fontes)

Diferença de caminho:
$$
\Delta r = r_2 - r_1
$$

Construtiva:
$$
\Delta r = m\lambda, \quad m \in \mathbb{Z}
$$

Destrutiva:
$$
\Delta r = \left(m + \frac{1}{2}\right)\lambda
$$
