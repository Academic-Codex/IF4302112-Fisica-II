# Fenômenos Ondulatórios

## Intensidade e Amplitude

A intensidade de uma onda é proporcional ao quadrado da amplitude:

$$
I \propto A^2
$$

---

## Ondas Transversais vs Longitudinais

**Ondas transversais:**  
A perturbação é perpendicular à direção de propagação.

**Ondas longitudinais:**  
A perturbação ocorre na mesma direção da propagação.

---

# Equação da Onda Longitudinal (Modelo Discreto)

## Sistema relaxado

A posição do $m$-ésimo átomo é

$$
x_m = m \ell_0, \quad m \in \mathbb{N}
$$

## Sistema deformado

$$
x_m^{(p)} = m \ell_0 + u_m = x_m + u_m
$$

onde $u_m$ é o deslocamento em relação à posição de equilíbrio.

---

## Forças resultantes no $m$-ésimo átomo

A força total é dada por

$$
F_m = K_{mi}(u_{m+1} - u_m) - K_{mi}(u_m - u_{m-1})
$$

ou, equivalentemente,

$$
F_m = K_{mi} \big[ (u_{m+1} - u_m) - (u_m - u_{m-1}) \big]
$$

---

# Equação da Onda 

No limite contínuo, obtemos a equação da onda:

$$
\frac{\partial^2 u}{\partial x^2}
=
\frac{1}{v^2}
\frac{\partial^2 u}{\partial t^2}
$$

onde

$$
v = \sqrt{\frac{Y}{\rho}}
$$

---

## Elementos de uma onda harmônica

$$
u(x,t) = A \cos(kx - \omega t + \varphi)
$$

onde:

- $A$ é a amplitude  
- $k$ é o número de onda  
- $\omega$ é a frequência angular  
- $\varphi$ é a defasagem  

Número de onda:

$$
k = \frac{2\pi}{\lambda}
$$

Frequência angular:

$$
\omega = 2\pi f = \frac{2\pi}{T} = kv
$$

Velocidade de propagação:

$$
v = \lambda f = \frac{\omega}{k}
$$

**Observação:**

- $kx - \omega t$: onda se propaga no sentido positivo de $x$  
- $kx + \omega t$: onda se propaga no sentido negativo de $x$

---

## Ondas transversais — forças

A força resultante pode ser escrita como

$$
F_R = F_D \sin\theta_m - F_E \sin\theta_{m+1}
$$

No limite de pequenas deformações:

$$
F_R = K_m \frac{d - d_0}{d}
\left[
\frac{u_{m+1} - u_m}{d}
-
\frac{u_m - u_{m-1}}{d}
\right]
$$

No limite contínuo:

$$
d\,K_m (d - d_0)\,\frac{\partial^2 u}{\partial x^2}
=
m\,\frac{\partial^2 u}{\partial t^2}
$$

---

## Força de tensão e densidade linear

A força de tensão é

$$
F_T = K(\Delta d)
$$

Definindo a densidade linear:

$$
\mu = \frac{m}{d}
$$

A equação da onda fica:

$$
\frac{\partial^2 u}{\partial x^2}
=
\frac{1}{v^2}
\frac{\partial^2 u}{\partial t^2}
$$

com

$$
v = \sqrt{\frac{F_T}{\mu}}
$$

---

## Propagação de energia

### Visão microscópica

Na região não perturbada da corda, as partículas já possuem energia cinética
(agitação térmica) e energia potencial (ligações interatômicas).

A onda transporta energia ao aumentar a energia das partículas envolvidas.

### Visão macroscópica

Na região não perturbada, o elemento da corda está em repouso e possui energia
potencial constante associada à tensão.

Quando o pulso chega, o elemento se deforma e sua energia aumenta.

---

## Energia cinética (densidade)

$$
\Delta K = \frac{1}{2} m v^2
$$

Como $v = \frac{\partial u}{\partial t}$ e $m = \mu\,\Delta x$:

$$
\Delta K
=
\frac{1}{2}\,\mu
\left(\frac{\partial u}{\partial t}\right)^2
\Delta x
$$

---

## Energia potencial (densidade)

$$
\Delta U = \frac{1}{2} K_{\text{mola}} (\ell - \Delta x)^2
$$

Aproximando para pequenas deformações:

$$
\ell \approx \Delta x \sqrt{1 + \left(\frac{\partial u}{\partial x}\right)^2}
$$

Logo:

$$
\Delta U
=
\frac{1}{2} F_T
\left(\frac{\partial u}{\partial x}\right)^2
\Delta x
$$

---

## Energia mecânica total

Solução geral da equação da onda:

$$
u(x,t) = f(x - vt)
$$

Derivadas:

$$
\frac{\partial f}{\partial t} = -v f'
\quad\text{e}\quad
\frac{\partial f}{\partial x} = f'
$$

Assim:

$$
\left(\frac{\partial u}{\partial t}\right)^2
=
v^2
\left(\frac{\partial u}{\partial x}\right)^2
$$

Com isso:

$$
\Delta K
=
\frac{1}{2}\,\mu v^2
\left(\frac{\partial u}{\partial x}\right)^2
\Delta x
=
\frac{1}{2} F_T
\left(\frac{\partial u}{\partial x}\right)^2
\Delta x
=
\Delta U
$$

Energia total:

$$
\Delta E = \Delta K + \Delta U
=
F_T
\left(\frac{\partial u}{\partial x}\right)^2
\Delta x
$$

Para ondas harmônicas:

$$
\Delta E
=
\frac{1}{2}\,\mu \omega^2 A^2
$$

---

## Observação importante

A energia de um oscilador qualquer é

$$
E = \frac{1}{2} k A^2
$$

---

## Caso específico: ondas harmônicas

$$
u(x,t) = A \cos(kx - \omega t)
$$

Energia cinética total em um comprimento de onda:

$$
K
=
\int_0^\lambda
\frac{1}{2}\,\mu
\left(\frac{\partial u}{\partial t}\right)^2
dx
=
\frac{1}{4}\,\mu \omega^2 A^2 \lambda
$$

Energia potencial total:

$$
U
=
\int_0^\lambda
\frac{1}{2} F_T
\left(\frac{\partial u}{\partial x}\right)^2
dx
=
\frac{1}{4}\,\mu \omega^2 A^2 \lambda
$$

---

## Potência média

$$
\bar{P}
=
\frac{K + U}{T}
=
\frac{1}{2}\,\mu \omega^2 A^2 v
$$

A potência média **não depende** do comprimento da corda.

---

## Potência média por área transversal

$$
\bar{P}_{\text{área}} = \frac{\bar{P}}{A_{\text{transversal}}}
$$
