# Máquinas Térmicas

## Caminhos / Circuitos Fechados

Entropia e energia interna ($E_{\text{int}}$) são **funções de estado**:  
dependem apenas dos estados inicial e final de um processo térmico,  
e **não** do caminho termodinâmico percorrido.

Trabalho ($W$) e calor ($Q$) **não são funções de estado**.  
O trabalho pode ser interpretado como a **área interna de um circuito fechado**
no diagrama $P$–$V$.

Se
$$
(T_i, V_i, P_i) = (T_f, V_f, P_f),
$$
então
$$
\Delta E_{\text{int}} = \Delta S = 0.
$$

A área $A$ do ciclo no diagrama $P$–$V$ satisfaz:
$$
A = W.
$$

---

## Definições Importantes

### Processos reversíveis

Processos termodinâmicos **quase-estáticos**, que permitem caminhos bem definidos
no diagrama $P$–$V$.

**Exemplos:**
- Isotérmico  
- Adiabático  
- Isocórico  
- Isobárico  

---

### Máquinas térmicas

Sistema que:
1. retira calor de uma **fonte quente**,
2. transforma parte desse calor em **trabalho mecânico**,
3. devolve o restante a uma **fonte fria**.

São representadas por **ciclos** em diagramas $P$–$V$, normalmente formados por
processos reversíveis.

---

### Máquina térmica reversível

Máquina térmica ideal que **não aumenta a entropia do universo**.

O único ciclo capaz de representar uma máquina térmica reversível é o  
**Ciclo de Carnot**.

---

## Ciclo de Carnot

Etapas do ciclo:

1. Expansão isotérmica à temperatura $T_A$  
2. Expansão adiabática  
3. Compressão isotérmica à temperatura $T_B$  
4. Compressão adiabática  

---

## Eficiência de uma Máquina Térmica (Carnot)

$$
\eta = \frac{W}{Q_A}
     = \frac{Q_A - Q_B}{Q_A}
     = 1 - \frac{Q_B}{Q_A}
     = 1 - \frac{T_B}{T_A}.
$$

- Fonte quente: $A$  
- Fonte fria: $B$

A relação
$$
\frac{Q_A}{T_A} = \frac{Q_B}{T_B}
$$
vale **apenas** para o ciclo de Carnot.

---

## Eficiência de um Refrigerador (Carnot)

$$
\eta = \frac{Q_B}{W}
     = \frac{Q_B}{Q_A - Q_B}
     = \frac{1}{\frac{Q_A}{Q_B} - 1}
     = \frac{1}{\frac{T_A}{T_B} - 1}.
$$

---

## Eficiência de uma Bomba de Calor (Carnot)

$$
\eta = \frac{Q_A}{W}
     = \frac{Q_A}{Q_A - Q_B}
     = \frac{1}{1 - \frac{Q_B}{Q_A}}
     = \frac{1}{1 - \frac{T_B}{T_A}}.
$$

---

## Variação de Entropia em Transições de Fase

Durante uma mudança de fase:

$$
\Delta S = \int \frac{dQ_R}{T}
         = \frac{\Delta Q_R}{T}
         = \frac{m\,L}{T},
$$

onde:
- $L$ é o **calor latente**,
- $m$ é a massa do sistema.

---

## Teorema de Clausius

Em um ciclo reversível:

$$
\oint \frac{dQ}{T} = 0.
$$

Onde:
- $dQ$ é o calor recebido pelo fluido em cada etapa;
- $T$ é a temperatura do meio externo com o qual o calor é trocado.
