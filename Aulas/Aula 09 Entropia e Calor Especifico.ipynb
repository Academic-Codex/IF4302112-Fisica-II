# Calor específico do sólido de Einstein  

## Calor específico do sólido de Einstein

O calor específico **por átomo** é definido como

$$
c_{\text{átomo}} = \frac{\Delta E}{\Delta T}\,\frac{1}{N_{\text{átomos}}}
$$

ou, de forma discreta,

$$
c_{\text{átomo}}
= \frac{E(i+1) - E(i-1)}{T^{+} - T^{-}}\,
\frac{1}{N_{\text{átomos}}}
$$

### Definições de temperatura

Temperatura à direita:

$$
T^{+} = \frac{E(i+1) - E(i)}{S(i+1) - S(i)}
$$

Temperatura à esquerda:

$$
T^{-} = \frac{E(i) - E(i-1)}{S(i) - S(i-1)}
$$

Temperatura média:

$$
T_{\text{média}} = \frac{T^{+} + T^{-}}{2}
= \frac{E(i+1) - E(i-1)}{S(i+1) - S(i-1)}
$$

### Observação

Para o sólido de Einstein, a constante elástica $K_s$ é **quatro vezes maior**
que a constante elástica associada ao módulo de Young:

$$
4\,K_s^{(\text{Young})} = K_s^{(\text{Einstein})}
$$

---

## Energia de um quantum no sólido de Einstein

A energia de um quantum é

$$
\varepsilon = \hbar \omega_0
$$

com

$$
\omega_0 = \sqrt{\frac{K_s^{(\text{Einstein})}}{m_{\text{atômica}}}}
$$

Logo,

$$
\varepsilon
= \hbar \sqrt{\frac{K_s^{(\text{Einstein})}}{m_{\text{atômica}}}}
= \hbar \sqrt{\frac{4K_s^{(\text{Young})}}{m_{\text{molar}}/N_A}}
$$

### Constantes

Constante de Planck reduzida:

$$
\hbar = 1{,}05 \times 10^{-34}\ \text{J}\cdot\text{s}
$$

Número de Avogadro:

$$
N_A = 6{,}022 \times 10^{23}
$$
