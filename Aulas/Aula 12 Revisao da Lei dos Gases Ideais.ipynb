## Versão macroscópica da lei dos gases ideais

A equação de estado é:
$$
pV = N R T
$$

onde:
- $N$ é o número de mols  
- $R \simeq 8{,}3\ \text{J mol}^{-1}\text{K}^{-1}$ é a **constante dos gases ideais**
- $T$ é a temperatura absoluta

---

## Trabalho realizado por um pistão sobre um gás

O trabalho é dado por:
$$
W = - \int_{V_1}^{V_2} p_{\text{ext}}\, dV
$$

---

## Primeira Lei da Termodinâmica
$$
\Delta E_{\text{int}} = W + Q
$$

**Observação:**  
Se a compressão de um gás ideal é **isotérmica** (temperatura constante), então:
$$
\Delta E_{\text{int}} = 0 \;\;\Rightarrow\;\; W + Q = 0
$$

---

## Capacidade térmica

### Volume constante

A variação da energia térmica é:
$$
\Delta E_{\text{térmica}} = N\, C_V\, \Delta T
$$

onde:
- $C_V$ é a **capacidade térmica específica a volume constante**

# Entropia, equação de estado e máquinas térmicas

# Equação dos gases ideais (1 mol)
$$
P V = R T
$$

Derivando:
$$
P\,dV + V\,dP = R\,dT
$$

Podemos representar a variação da entropia em função de diferentes grandezas.

# (a) Entropia como função de (V, T)
$$
dS = \frac{C_V(T)}{T}\,dT + R\,\frac{dV}{V}
$$

Integrando entre os estados inicial e final:
$$
\Delta S
= \int_{T_i}^{T_f} \frac{C_V(T)}{T}\,dT
+ R \int_{V_i}^{V_f} \frac{dV}{V}
$$

Se $C_V$ é constante:
$$
\Delta S = C_V \ln\!\left(\frac{T_f}{T_i}\right)
+ R \ln\!\left(\frac{V_f}{V_i}\right)
$$

# (b) Entropia como função de (P, T)
Usando $P\,dV = V\,dP + R\,dT$:

$$
dS = \frac{C_V + R}{T}\,dT - R\,\frac{dP}{P}
$$

Como $C_P = C_V + R$:
$$
dS = \frac{C_P}{T}\,dT - R\,\frac{dP}{P}
$$

Integrando:
$$
\Delta S
= C_P \ln\!\left(\frac{T_f}{T_i}\right)
- R \ln\!\left(\frac{P_f}{P_i}\right)
$$

# (c) Entropia como função de (P, V)
$$
\Delta S
= C_V \ln\!\left(\frac{P_f V_f^{\gamma}}{P_i V_i^{\gamma}}\right),
\qquad
\gamma = \frac{C_P}{C_V}
$$

# Observação sobre número de mols
Todos os resultados acima são para 1 mol.
Para $n$ mols:
$$
\Delta S(n) = n\,\Delta S(1\,\text{mol})
$$
