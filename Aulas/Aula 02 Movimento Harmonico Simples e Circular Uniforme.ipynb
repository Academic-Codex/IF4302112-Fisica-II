# Movimento harmônico simples e movimento circular uniforme

## Notação complexa

### Características de um número complexo

$$
z = a + i b
$$

$$
\Re(z) = a
$$

$$
\Im(z) = b
$$

$$
|z|^2 = a^2 + b^2
$$

$$
|z| = \sqrt{a^2 + b^2}
$$

---

## Fórmula de Euler

$$
e^{i x} = \cos x + i \sin x
$$

$$
\Re\left(e^{i x}\right) = \cos x = \frac{e^{i x} + e^{-i x}}{2}
$$

$$
\Im\left(e^{i x}\right) = \sin x = \frac{e^{i x} - e^{-i x}}{2 i}
$$

---

A imagem $z = x + i y$ do número complexo é um ponto de coordenadas cartesianas $(x,y)$.  
Se passarmos para coordenadas polares, temos $(r,\theta)$:

$$
x = r \cos \theta
$$

$$
y = r \sin \theta
$$

$$
z = x + i y = r(\cos \theta + i \sin \theta) = r e^{i \theta}
$$

O argumento de $z$ $(\arg z)$ é dado por:

$$
\arg z = \theta = \arctan\left(\frac{y}{x}\right)
$$

---

## Aplicação

$$
\ddot{z} + \omega^2 z = 0
$$

A solução complexa dessa equação é:

$$
z(t) = A e^{i(\omega t + \psi)}
$$

onde:

$$
\Re[z(t)] = x(t) = A \cos(\omega t + \psi)
$$

### Obs.

A operação de derivação de $z$ é igual a multiplicar por $i\omega$ para a função $z(t)$:

$$
\frac{d}{dt} \;\longleftrightarrow\; i \omega
$$

Por exemplo:

$$
\frac{dz}{dt} = i \omega A e^{i(\omega t + \psi)}
$$

$$
\frac{dx}{dt} = \Re\left(\frac{dz}{dt}\right) = -\omega A \sin(\omega t + \psi)
$$

# Oscilações amortecidas

## Oscilações amortecidas e forçadas

A resistência de um fluido, como o ar, ao deslocamento de um corpo é proporcional à velocidade para velocidades suficientemente pequenas, o que se aplica a pequenas oscilações.

Vamos considerar uma força de amortecimento proporcional à velocidade.  
Com isso temos:

$$
m \ddot{x} = -k x - p \dot{x}, \quad p > 0
$$

onde $(-p \dot{x})$ representa a resistência dissipativa, que atua em sentido oposto à velocidade.

Dividindo por $m$ os dois membros, obtemos:

$$
\ddot{x} + \gamma \dot{x} + \omega_0^2 x = 0
$$

onde:

$$
\omega_0^2 = \frac{k}{m}
$$

$$
\gamma = \frac{p}{m}
$$

---

## Tipos de oscilador amortecido

### Subcrítico

Se $\frac{\gamma}{2} < \omega_0$ (mais demorado para voltar à posição de equilíbrio):

$$
x(t) = A e^{-\frac{\gamma}{2} t} \cos(\omega t + \phi)
$$

sendo:

$$
\omega = \sqrt{\omega_0^2 - \left(\frac{\gamma}{2}\right)^2}
$$

---

### Crítico

Se $\frac{\gamma}{2} = \omega_0$ (mais rápido):

$$
x(t) = e^{-\frac{\gamma}{2} t}(A + B t)
$$

---

### Supercrítico

Se $\frac{\gamma}{2} > \omega_0$ (intermediário):

$$
x(t) = e^{-\frac{\gamma}{2} t}\left(A e^{-\beta t} + B e^{\beta t}\right)
$$

onde:

$$
\beta = \sqrt{\left(\frac{\gamma}{2}\right)^2 - \omega_0^2}
$$

---

## Condições iniciais e características

### Subcrítico

$$
x(t) = A e^{-\frac{\gamma}{2} t} \cos(\omega t + \phi)
$$

$$
v(t) = -A e^{-\frac{\gamma}{2} t}
\left(
\frac{\gamma}{2} \cos(\omega t + \phi)
+ \omega \sin(\omega t + \phi)
\right)
$$

O período é:

$$
T = \frac{2\pi}{\omega}
$$

ainda vale.

---

## Amplitude no regime amortecido

A amplitude decai segundo:

$$
A(t) = A_0 e^{-\frac{\gamma}{2} t}
$$

---

## Caso crítico

$$
x(t) = e^{-\frac{\gamma}{2} t}(A + B t)
$$

com:

$$
A = x_0
$$

$$
B = v_0 + \frac{\gamma}{2} x_0
$$

Dos três tipos de amortecimento, o **crítico** é aquele que retorna ao ponto de equilíbrio **mais rapidamente**.

---

## Supercrítico (forma equivalente)

$$
x(t) = e^{-\frac{\gamma}{2} t}
\left(
A e^{-\beta t} + B e^{\beta t}
\right)
$$

$$
= A e^{-t\left(\frac{\gamma}{2} + \beta\right)} + B e^{-t\left(\frac{\gamma}{2} - \beta\right)}
$$

# O balanço de energia

$$
E(t) = \frac{1}{2} m \dot{x}^2(t) + \frac{1}{2} k x^2(t)
$$

Note que a energia **não é conservativa**; a taxa de variação da energia é

$$
\frac{dE}{dt}
= m \ddot{x}\dot{x} + k x \dot{x}
= \dot{x}(m\ddot{x} + kx)
$$

Usando a equação de movimento amortecida:

$$
m\ddot{x} + kx = -p\dot{x}
$$

obtemos

$$
\frac{dE}{dt} = -p \dot{x}^2 = -m\gamma \dot{x}^2
$$

que representa a **potência dissipada**.

---

## Fator de mérito $Q$

$$
Q = 2\pi
\left(
\frac{\text{energia armazenada}}
{\text{energia dissipada por ciclo}}
\right)
$$

$$
Q = \frac{2\pi E}{\Delta E}
= \frac{2\pi}{\gamma T}
= \frac{\omega_0}{\gamma}
$$

A energia média é

$$
\bar{E}
= \frac{1}{2} m \omega_0^2 A^2 e^{-\gamma t}
$$

e satisfaz

$$
\frac{d\bar{E}}{dt} = -\gamma \bar{E}
$$

**Obs.:** No gráfico de energia, quando a derivada no ponto é zero, naquele ponto a energia cinética é nula e a energia potencial é igual à energia total.

---

# Oscilações forçadas

$$
\ddot{x} + \gamma \dot{x} + \omega_0^2 x
= \frac{F_{\text{ext}}(t)}{m}
$$

A solução geral é

$$
x(t) = x_h(t) + x_p(t)
$$

onde:

- $x_h(t)$ é a solução homogênea (subcrítica / crítica / supercrítica);
- $x_p(t)$ é a solução particular (regime estacionário).

---

## Caso de força harmônica

$$
\ddot{x} + \gamma \dot{x} + \omega_0^2 x
= \frac{F_0}{m}\cos(\omega_e t)
$$

A solução particular é

$$
x_p(t) = A(\omega_e)\cos(\omega_e t + \psi(\omega_e))
$$

com

$$
A(\omega_e)
= \frac{F_0/m}
{\sqrt{(\omega_0^2 - \omega_e^2)^2 + (\gamma \omega_e)^2}}
$$

$$
\psi(\omega_e)
= -\arctan\left(
\frac{\gamma \omega_e}{\omega_0^2 - \omega_e^2}
\right)
$$

onde $\omega_e$ é a **frequência angular externa** e

$$
T_e = \frac{2\pi}{\omega_e}
$$

é o **período externo**.

---

## Relação entre as frequências

### Limite de baixas frequências ($\omega_e \ll \omega_0$)

O movimento é dominado pela **força restauradora**.

---

### Limite de altas frequências ($\omega_e \gg \omega_0$)

O deslocamento está em **oposição de fase** com a força externa:

$$
\psi = -\pi
$$

O movimento é dominado pela **inércia**.

---

### Ressonância ($\omega_e \approx \omega_0$)

$$
A \to \infty \quad \text{se } \gamma \to 0
$$

Para $\gamma \ll \omega_0$:

$$
A_{\max}
= \frac{F_0}{m\omega_0\gamma}
$$

O fator de qualidade é

$$
Q = \frac{\omega_0}{\gamma}
$$

---

# Balanço de energia (forçado)

Para oscilações amortecidas forçadas:

$$
\frac{dE}{dt}
= -m\gamma \dot{x}^2 + P(t)
$$

onde $P(t)$ é a potência fornecida pela força externa:

$$
P(t) = F(t)\dot{x}(t)
$$

---

## Potência no regime estacionário (força harmônica)

$$
F(t) = F_0\cos(\omega t)
$$

$$
\dot{x}(t) = -\omega A \sin(\omega t + \psi)
$$

Combinando os termos:

$$
\frac{dE}{dt}
= \frac{1}{2} m\omega(\omega^2 - \omega_0^2)A^2
\sin\left[2(\omega t + \psi)\right]
$$

Mostra-se que a energia instantânea oscila com frequência $2\omega$.

A energia média satisfaz:

$$
\left\langle \frac{dE}{dt} \right\rangle = 0
$$

como esperado no regime estacionário.

---

## Potência média dissipada

$$
\bar{P} = m\gamma \langle \dot{x}^2 \rangle
$$

Como

$$
\dot{x}^2
= \omega^2 A^2 \sin^2(\omega t + \psi)
= \frac{1}{2}\omega^2 A^2
$$

segue que

$$
\bar{P}(\omega_e)
= \frac{1}{2}\gamma m\omega_e^2 A^2
= \frac{\gamma F_0^2 \omega_e^2}
{2m\left[(\omega_e^2-\omega_0^2)^2 + \gamma^2\omega_e^2\right]}
$$

A potência média máxima ocorre na ressonância $\omega_e=\omega_0$:

$$
\bar{P}_{\max}
= \frac{F_0^2}{2m\omega_0\gamma}
$$

Para $Q \gg 1$ e $\omega_e \approx \omega_0$:

$$
\bar{P}(\omega_e)
\simeq
\frac{\gamma F_0^2}{8m}
\frac{1}{(\omega_e-\omega_0)^2 + \gamma^2/4}
$$

---

## Forma alternativa da potência média

$$
\bar{P}(\omega_e)
= -\frac{\omega_e F_0 A}{2}\sin\psi
$$

O fator $|\sin\psi|$ é chamado **fator de potência**.

A potência é máxima na ressonância, quando

$$
\psi = -\frac{\pi}{2}
$$

---

## Energia média armazenada no oscilador

$$
\bar{E}
= \frac{1}{4} m(\omega_e^2 + \omega_0^2)A^2
$$

A energia média dissipada durante um período $T$ é

$$
\Delta E
= \bar{P}T
= \frac{2\pi}{\omega_e}\,\bar{P}
$$

---

## Fator $Q$ do oscilador forçado amortecido

$$
Q
= 2\pi
\left(
\frac{\text{energia média armazenada}}
{\text{energia média perdida por ciclo}}
\right)
$$

$$
Q
= \frac{1}{2}
\left(
\frac{\omega_e}{\omega_0}
+ \frac{\omega_0}{\omega_e}
\right)
\frac{\omega_0}{\gamma}
$$
