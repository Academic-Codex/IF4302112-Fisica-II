# Tensão, deformação e módulo de Young

## Deformação específica

$$
\text{deformação específica}
= \frac{\Delta L}{L}
$$

---

## Tensão

$$
\text{tensão}
= \frac{\text{força de tração}}{\text{área da seção transversal}}
= \frac{F_T}{A}
$$

---

## Módulo de Young

$$
Y
= \frac{\text{tensão}}{\text{deformação}}
= \frac{F_T/A}{\Delta L/L}
$$

O módulo de Young é uma **propriedade do material**, independente da forma ou do tamanho do objeto.  
Quanto mais duro o material, maior é $Y$.

---

## Forma alternativa

$$
\frac{F_T}{A} = Y\frac{\Delta L}{L}
$$

Em nível interatômico:

$$
Y = \frac{k_m}{d}
$$

onde:

- $k_m$ é a rigidez de uma ligação interatômica;
- $d$ é o comprimento da ligação (diâmetro atômico).

A massa de um átomo é

$$
m_{\text{átomo}}
= \frac{M_{\text{molar}}}{N_A}
$$

com $N_A \approx 6\times 10^{23}$.
