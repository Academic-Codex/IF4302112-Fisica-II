# Exemplos e aplicações

## a) O pêndulo de torção

Consideremos uma barra horizontal suspensa em equilíbrio por um fio vertical.  
Se defletirmos (movermos) a barra no plano horizontal de um pequeno ângulo em relação à posição de equilíbrio, a lei de Hooke para a torção do fio diz que ele reage com um torque restaurador proporcional ao ângulo de torção:

$$
\tau = -k \, \phi
$$

$k$ é o módulo de torção, que depende do seu diâmetro, comprimento e material.

Analisando a fórmula do torque $(\tau = I \alpha)$, temos:

$$
I \ddot{\phi} = -k \phi
$$

ou

$$
\ddot{\phi} + \omega^2 \phi = 0
$$

onde

$$
\omega^2 = \frac{k}{I}
$$

## b) O pêndulo simples

A equação do movimento angular é:

$$
m a_\theta = m l \frac{d^2 \theta}{dt^2} = - m g \sin \theta
$$

Logo,

$$
\frac{d^2 \theta}{dt^2} = -\frac{g}{l} \sin \theta
$$

Lembrando que $a_\theta = a_R$.

Aproximando $\sin \theta \approx \theta$, aproximação que só pode ser feita para pequenos ângulos, temos:

$$
\ddot{\theta} + \frac{g}{l} \theta = 0
$$

ou

$$
\ddot{\theta} + \omega^2 \theta = 0
$$

onde

$$
\omega^2 = \frac{g}{l}
$$

O período para pequenas oscilações do pêndulo é:

$$
T = 2\pi \sqrt{\frac{l}{g}}
$$

Obs.: $T$ é independente da amplitude de oscilação.

---

## Energia do pêndulo

A energia cinética é:

$$
K(t) = \frac{1}{2} m v^2 = \frac{1}{2} m l^2 \dot{\theta}^2
$$

A energia potencial (caso geral) é:

$$
U(t) = m g l (1 - \cos \theta)
$$

Para pequenos ângulos:

$$
U(t) = \frac{1}{2} m g l \theta^2 = \frac{1}{2} m \omega^2 l^2 \theta^2
$$

### Energia total

$$
E = \frac{1}{2} m l^2 \dot{\theta}^2 + m g l (1 - \cos \theta)
$$

### Posições de equilíbrio

O pêndulo estará em equilíbrio estável quando $\theta = 0$  
e instável quando $\theta = \pi$.

## c) O pêndulo físico

O torque é dado por:

$$
\tau = - M g s \sin \theta
$$

Logo,

$$
I \ddot{\theta} = - M g s \sin \theta
$$

O momento de inércia pode ser escrito como:

$$
I = M k^2
$$

com

$$
k^2 = l s
$$

e

$$
s_c = l - s
$$

## d) Oscilações de um líquido em tubos em U

$p$ = densidade  
$A$ = seção transversal  

A energia potencial é:

$$
U(y) = p A y \, g \, y = p A g y^2
$$

A energia cinética é:

$$
K(y) = \frac{1}{2} p A l \left(\frac{dy}{dt}\right)^2
$$

A energia total é:

$$
E = \frac{1}{2} p A l \dot{y}^2 + p A g y^2
$$

Comparando com a energia de um oscilador harmônico:

$$
\frac{1}{2} K = \frac{1}{2} M \omega^2
$$

com

$$
M = p A l
$$

Temos:

$$
p A g = \frac{1}{2} p A l \omega^2
$$

Logo,

$$
\omega^2 = \frac{2g}{l}
$$

Com isso é possível encontrar a frequência angular das oscilações harmônicas da massa líquida.

### Frase de Newton

> “Se a água sobe e desce alternadamente nos ramos de um pêndulo cujo comprimento entre o ponto de oscilação é igual à metade do comprimento da água no tubo, digo então que a água subirá e descerá nos mesmos tempos em que o pêndulo oscila.”

## e) Oscilações de duas partículas acopladas

$l$ = comprimento de equilíbrio da mola

Considerando:

$$
x = (x_2 - x_1) - l
$$

Temos que:

$$
F_1 = K x = -F_2
$$

$$
m_1 \ddot{x}_1 = K x
$$

$$
m_2 \ddot{x}_2 = -K x
$$

o que leva a:

$$
\mu \ddot{x} = -K x
$$

onde

$$
\mu = \frac{m_1 m_2}{m_1 + m_2}
$$

A energia total pode ser escrita como:

$$
E = E_{CM} + E'
$$

com

$$
E_{CM} = K_{CM} = \frac{M V^2}{2}
$$

e

$$
E' = \frac{1}{2} \mu \dot{x}^2 + \frac{1}{2} K x^2
$$

A energia do centro de massa e a energia interna se conservam separadamente;  
a translação do CM não afeta a oscilação.
