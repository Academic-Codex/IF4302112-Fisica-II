# Fenômenos Ondulatórios

## Intensidade e Amplitude

A intensidade de uma onda é proporcional ao quadrado da amplitude:

$$
I \propto A^2
$$

---

## Ondas Transversais vs Longitudinais

**Ondas transversais:**  
A perturbação é perpendicular à direção de propagação.

**Ondas longitudinais:**  
A perturbação ocorre na mesma direção da propagação.

---

# Equação da Onda Longitudinal (Modelo Discreto)

## Sistema relaxado

A posição do $m$-ésimo átomo é

$$
x_m = m \ell_0, \quad m \in \mathbb{N}
$$

## Sistema deformado

$$
x_m^{(p)} = m \ell_0 + u_m = x_m + u_m
$$

onde $u_m$ é o deslocamento em relação à posição de equilíbrio.

---

## Forças resultantes no $m$-ésimo átomo

A força total é dada por

$$
F_m = K_{mi}(u_{m+1} - u_m) - K_{mi}(u_m - u_{m-1})
$$

ou, equivalentemente,

$$
F_m = K_{mi} \big[ (u_{m+1} - u_m) - (u_m - u_{m-1}) \big]
$$

---

# Equação da Onda (Limite Contínuo)

No limite contínuo, obtemos a equação da onda:

$$
\frac{\partial^2 u}{\partial x^2}
=
\frac{1}{v^2}
\frac{\partial^2 u}{\partial t^2}
$$

onde

$$
v = \sqrt{\frac{Y}{\rho}}
$$

---

# Onda Harmônica

Uma solução harmônica da equação da onda é

$$
u(x,t) = A \cos(kx - \omega t + \varphi)
$$

onde:
- $A$ é a amplitude,
- $k$ é o número de onda,
- $\omega$ é a frequência angular,
- $\varphi$ é a fase inicial.

---

## Parâmetros da Onda

Número de onda:

$$
k = \frac{2\pi}{\lambda}
$$

Frequência angular:

$$
\omega = 2\pi f = \frac{2\pi}{T}
$$

Relação de dispersão:

$$
\omega = k v
$$

Logo,

$$
v = \frac{\omega}{k} = \lambda f
$$

---

**Observação:**  
- O sinal $kx - \omega t$ representa uma onda que se propaga no sentido positivo de $x$.
- O sinal $kx + \omega t$ representa propagação no sentido negativo de $x$.
